```python
from flask import Flask, render_template, request,jsonify
from flask_cors import CORS,cross_origin
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
```

```python
flipkart_url = "https://www.flipkart.com/search?q=" + 'iphonese'
```

```python
flipkart_url
```

```python
### Searching url
uClient = uReq(flipkart_url)
```

```python
### Http response
uClient
```

```python
### Reading Url client
flipkartPage = uClient.read()
```

```python
### This basiscally returns everything of Html pagefor the particular search
flipkartPage
```

```python
## Closing uClient
uClient.close()
```

```python
### parsing Entire Html using beautiful soap
### It can be little structured than raw data
flipkart_html = bs(flipkartPage, "html.parser")
```

```python
flipkart_html
```

```python
## Finding Search Url
### Getting Class id for the product which we are going to consider.
### This is a list
bigboxes= flipkart_html.findAll('div', {"class":"_1AtVbE col-12-12"})
```
